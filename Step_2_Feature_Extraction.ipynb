```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import AllChem, Descriptors
import warnings
warnings.filterwarnings('ignore')

print("="*70)
print("STEP 2: FEATURE EXTRACTION")
print("="*70)
```

```python
# Load merged data
print("\n[1/4] Loading data...")
df = pd.read_excel('Data_Final_merged.xlsx')
print(f"Dataset shape: {df.shape}")

smiles_list = df['SMILES_acc'].tolist()
print(f"Number of molecules: {len(smiles_list)}")
```

```python
# Generate Morgan Fingerprints
print("\n[2/4] Generating Morgan Fingerprints (radius=2, nBits=2048)...")
morgan_fps = []
failed_indices = []

for idx, smiles in enumerate(smiles_list):
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
        morgan_fps.append(np.array(fp))
    else:
        failed_indices.append(idx)
        morgan_fps.append(np.zeros(2048))

morgan_fps = np.array(morgan_fps)
print(f"✓ Morgan fingerprints shape: {morgan_fps.shape}")
print(f"  Failed conversions: {len(failed_indices)}")
```

```python
# Generate MACCS Keys (alternative features)
print("\n[3/4] Generating MACCS Keys (166 structural features)...")
maccs_fps = []

for idx, smiles in enumerate(smiles_list):
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        fp = AllChem.GetMACCSKeysFingerprint(mol)
        maccs_fps.append(np.array(fp))
    else:
        maccs_fps.append(np.zeros(167))

maccs_fps = np.array(maccs_fps)
print(f"✓ MACCS keys shape: {maccs_fps.shape}")
```

```python
# Save features
print("\n[4/4] Saving feature matrices...")
np.save('features_morgan.npy', morgan_fps)
np.save('features_maccs.npy', maccs_fps)

print("✓ Saved: features_morgan.npy")
print("✓ Saved: features_maccs.npy")
```

```python
# Feature statistics
print("\n" + "="*70)
print("FEATURE STATISTICS")
print("="*70)
print(f"Morgan Fingerprints:")
print(f"  Shape: {morgan_fps.shape}")
print(f"  Sparsity: {(morgan_fps == 0).sum() / morgan_fps.size * 100:.1f}%")
print(f"  Active bits per molecule (avg): {morgan_fps.sum(axis=1).mean():.1f}")

print(f"\nMACCS Keys:")
print(f"  Shape: {maccs_fps.shape}")
print(f"  Sparsity: {(maccs_fps == 0).sum() / maccs_fps.size * 100:.1f}%")
print(f"  Active keys per molecule (avg): {maccs_fps.sum(axis=1).mean():.1f}")
```
